1. Write a function that takes start and end of a range returns a Pandas series object containing numbers within that range. 

In case the user does not pass start or end or both they should default to 1 and 10 respectively. 

eg. 

 range_series() -> Should Return a pandas series from 1 to 10

 range_series(5) -> Should Return a pandas series from 5 to 10 

 range_series(5, 10) -> Should Return a pandas series from 5 to 10. 

```python
import numpy as np
import pandas as pd
```

```python
num=input("please enter numbers less than or equal to 10 seperated by comma: ")
print("num is: ",num)
if not num:
  print("user passed empty input")
  x = 1
  y = 10
else:
  tup=tuple(num.split(","))
  print("tup is: ",tup)
  print("len of tup:",len(tup))
  if len(tup)==1:
    x = int(tup[0])
    y = 10
  else:
    x = int(tup[0])
    y = int(tup[1])
  
print("x is:",x)
print("y is:",y)

def obj(x,y):
  L1=[]
  for p in range(x,y+1):
    print("p is: ",p)
    L1.append(p)
    print(L1)
  arr = np.array(L1)
  print("array is:",arr)
  ser = pd.Series(arr)
  print("series is: ",ser)

obj(x,y)
```

E.G: get_max_correlated_column('mpg') -> should return 'drat'

```python
df = pd.read_csv("cars.csv")
df.head()
```

```python
cars1 = df.iloc[:,2:]
cars1.head()
```

```python
correlation=cars1.corr()
print(correlation)

cor = pd.DataFrame(correlation)

col = input("Please enter the column name: ")

def maxCorr(col):
  df2 = cor[col]
  #print(df2)
  df3 = df2.sort_values(ascending=False)
  corrIndex = list(df3.index)
  print(" The column ", col , "is having maximum correlation with the column ", corrIndex[1])

maxCorr(col)




```
