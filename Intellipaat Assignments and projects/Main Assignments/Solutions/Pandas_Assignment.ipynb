```python
#A
Data = pd.read_csv("Tips.csv")

#if 3 columns needs to be extracted seperately
df1 = Data['Time']
df2 = Data['TotalBill']
df3 = Data['Tips']
```

```python
#B
Data1 = pd.read_excel("Tips1.xlsx")
Data1.head()

#merging two dataframes Data and Data1 on specific unique column

Data.merge(Data1,on="SINO")
```

```python
#C
#I am assuming you wanted to copy Data2 from Data instead of Data1 , 
# as Data1 (Tips.xlsx) contains only SINO,GENDER columns

Data2 = Data.copy()
Data2.head()

#total tips received across days
Data2.groupby('Day').agg({'Tips':"sum"})
```
