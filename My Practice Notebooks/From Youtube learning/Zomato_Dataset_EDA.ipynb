```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_country.columns
```

```python
df_country.shape
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
#to get in descending order
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# pie chart to show top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')

plt.show()
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating counts'})
```

```python
ratings
```

```python
# observations
# 1) when rating is b/w 4.5 to 4.9 --> Excellent
# 2) when rating is b/w 4.0 to 4.4 --> very good
# 3) when rating is b/w 3.5 to 3.9 -->  good
# 4) when rating is b/w 2.5 to 3.9 --> average
# 5) when rating is b/w 0.1 to 2.4 --> poor
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating',y='Rating counts',data=ratings)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating',y='Rating counts',hue = 'Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
#observations
# 1) not rated count is very high
# 2) max no of ratings are b/w 2.5 to 3.4
```

```python
#count plot
plt.figure(figsize=(12,8))
sns.countplot(x= 'Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# countries that gave Zero ratings

final_df.loc[final_df['Aggregate rating']==0.0,'Country'].value_counts()
```
