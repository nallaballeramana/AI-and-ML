```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
scaler = StandardScaler()

std_data = scaler.fit_transform(data)

scaled_data = pd.DataFrame(std_data)
```

```python
scaled_data.head()
```

```python
cols = [
           'Cement',
           'Blast Furnace Slag',
           'Fly Ash',
           'Water',
           'Superplasticizer',
           'Coarse Aggregate',
           'Fine Aggregate',
           'Age',
           'Concrete compressive strength'
]
```

```python
scaled_data.columns = cols
scaled_data.head()
```

```python
# splitting the dataste to input and output columns
# output is Concrete compressive strength
# remaining are input columns

x = scaled_data.iloc[:,:-1]
y = scaled_data.iloc[:,-1:]
```

```python
x.head()
```

```python
y.head()
```

```python
# splitting data into test and train
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 1)
```

```python
x_test
```

```python
x_train
```

```python
y_test
```

```python
y_train
```

```python
reg_model = LinearRegression()
```

```python
reg_model.fit(x_train,y_train)
```

```python
#predicting using model with test data
y_pred = reg_model.predict(x_test)
```

```python
y_pred
```

```python
# calculating r2 score
r2_score(y_test,y_pred)
```
